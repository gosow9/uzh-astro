Gabriel's horn is a well-known example of a surface with infinite area but finite volume. One way to express it is:

$$
r = \frac{1}{z}, \quad z \geq 1
$$

in cylindrical coordinates $(r, \phi, z)$.

On the surface itself, one can adopt $(r, \phi)$ as a coordinate system. Show that the surface has the metric:

$$
ds^2 = \left(1 + \frac{1}{r^4}\right) dr^2 + r^2 d\phi^2
$$

Hence, argue that the distance between two circles at $r = r_1$ and $r = r_2$ on Gabriel's horn is **not** $r_2 - r_1$.


### Solution

1. **Metric Derivation:**

   In cylindrical coordinates, the general metric is:
   $$
   ds^2 = dr^2 + r^2 d\phi^2 + dz^2
   $$

   On Gabriel's horn, $ r = \frac{1}{z} $, so:
   $$
   dr = -\frac{1}{z^2} dz = -r^2 dz \implies dz = -\frac{1}{r^2} dr
   $$

   Substitute $ dz = -\frac{1}{r^2} dr $ into the metric:
   $$
   ds^2 = dr^2 + r^2 d\phi^2 + \left(-\frac{1}{r^2} dr\right)^2
   $$

   Simplify:
   $$
   ds^2 = dr^2 + r^2 d\phi^2 + \frac{1}{r^4} dr^2
   $$

   Combine terms:
   $$
   ds^2 = \left(1 + \frac{1}{r^4}\right) dr^2 + r^2 d\phi^2
   $$

2. **Distance Between Two Circles:**

   The radial distance between $ r_1 $ and $ r_2 $ is:
   $$
   \text{Distance} = \int_{r_1}^{r_2} \sqrt{1 + \frac{1}{r^4}} \, dr
   $$

   Since $ \sqrt{1 + \frac{1}{r^4}} > 1 $, the distance is greater than $ r_2 - r_1 $ so get a better understanding here the plot of the Horn with a distance path


```python
import numpy as np
import matplotlib.pyplot as plt

# Define Gabriel's Horn surface in cylindrical coordinates
def gabriel_horn(r):
    return 1 / r  # z = 1/r

r_values = np.linspace(0.5, 2, 500) 
z_values = gabriel_horn(r_values)

fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
theta = np.linspace(0, 2 * np.pi, 100)
R, Theta = np.meshgrid(r_values, theta)
Z = gabriel_horn(R)
X = R * np.cos(Theta)
Y = R * np.sin(Theta)

ax.plot_surface(X, Y, Z, alpha=0.7, rstride=10, cstride=10, cmap='viridis', edgecolor='none')

ax.plot(r_values[150:250], np.zeros_like(r_values[150:250]), z_values[150:250], color='red', lw=2, label='Integration Path from point a to b')

ax.set_title("Gabriel's Horn Surface with Integration Path")
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
ax.legend()

plt.show()
```
